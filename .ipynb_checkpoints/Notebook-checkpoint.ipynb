### Cerner le problème

- Le modèle peut être entraîné avec des exemples étiquetés (pour chaque district, les données sont fournies avec le résultat attendu, à savoir le prix médian des habitations dans ce district), il s'agit donc une tâche d'apprentissage supervisé. 
- Le système va utiliser plusieurs variables (telles que la population du district, le revenu médian, etc) pour effectuer une prédiction, il s'agit d'une régression multiple. C'est aussi un problème de régression univariée car nous tentons de prédire qu'une seule valeur pour chaque district.

```python
from pathlib import Path
import pandas as pd
import numpy as np
import tarfile
import urllib.request

def load_housing_data() :
    tarball_path = Path("datasets/housing.tgz")
    if not tarball_path.is_file() :
        Path("datasets").mkdir(parents=True, exist_ok=True)
        url = "https://github.com/ageron/data/raw/main/housing.tgz"
        urllib.request.urlretrieve(url, tarball_path)
        with tarfile.open(tarball_path) as housing_tarball:
            housing_tarball.extractall(path='datasets')
    return pd.read_csv(Path("datasets/housing/housing.csv"))

df = load_housing_data()
```

```python
strat_test_set['income_cat'].value_counts() / len(strat_test_set)
```

**Comparaison des biais d'échantionnage stratifié et d'échantionnage aléatoir**

```python
def income_cat_proportions(data):
    return data["income_cat"].value_counts() / len(data)

train_set, test_set = train_test_split(df, test_size=0.2, random_state=42)

compare_props = pd.DataFrame({
    "Overall %": income_cat_proportions(df),
    "Stratified %": income_cat_proportions(strat_test_set),
    "Random %": income_cat_proportions(test_set),
}).sort_index()
compare_props.index.name = "Income Category"
compare_props["Strat. Error %"] = (compare_props["Stratified %"] /
                                   compare_props["Overall %"] - 1)
compare_props["Rand. Error %"] = (compare_props["Random %"] /
                                  compare_props["Overall %"] - 1)
(compare_props * 100).round(2)
```

En comparant les proportions des différentes catégories de revenus dans le jeu de données tout entier, dans le jeu de test généré par `échantionnage stratifié` et dans le jeu de test généré par `échantionnage` purement aléatoire. Le jeu de test généré par `échantionnage stratifié` a des proportions de catégories de revenus à peu près identiques à celles du jeu de données tout entier, tandis que le jeu de test généré par un échantionnage purement aléatoire est faussé.

### 5.4 Transformateurs sur mesure

- Pour certaines opérations de nettoyage spécifiques ou pour combiner certaines variables.
- Un transformateur sur mesure utilisant l'algorithme KMeans pour identifier mes principaux clusters dans les données d'entraînement, puis utilise `rbf_kernel` dans la méthode `transform()` pour mesurer la disimilarité entre chaque échantillon et chaque centre de cluster.
- Le nombre de cluster recherchés est contrôlé par l'hyperparamètre `n_clusters`. Après l'entraînement, les centres des clusters peuvent être obtenus à partir de l'attribut `cluster_centers`

```python
from sklearn.cluster import KMeans
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import FunctionTransformer
from sklearn.metrics.pairwise import rbf_kernel
```

```python
class ClusterSimilarity(BaseEstimator, TransformerMixin) : 
    def __init__(self, n_clusters=10, gamma=1.0, random_state=None):
        self.n_clusters = n_clusters
        self.gamma = gamma
        self.random_state = random_state
    
    def fit(self, X, y=None, sample_weight=None):
        self.kmeans_ = KMeans(self.n_clusters, n_init=10, random_state=self.random_state)
        self.kmeans_.fit(X, sample_weight=sample_weight)
        return self
    
    def transform(self, X):
        return rbf_kernel(X, self.kmeans_.cluster_centers_, gamma=self.gamma)
    
    def get_features_names_out(self, names=None):
        return [f'Similarité au cluster {i}' for i in range(self.n_clusters)]
```

- Le transformateur `ClusterSimimarity` est créé, et fixe le nombre de clusters à 10. Puis il appelle `fit_transform()` en lui transmettant la latitude et la longitude de chaque district de jeu d'entrainement, en pondérant chaque district par la valeur moyenne des logements.
- Le transformateur utilise les k-moyennes pour localiser les clusters, puis mesure la similarité de type RBF gaussienne entre chacun des districts et les 10 centre de clusters.
- Le résultat est une matrice composant une ligne par district et une colonne par cluster.

```python
cluster_simil = ClusterSimilarity(n_clusters=10, gamma=1., random_state=42)
similarities = cluster_simil.fit_transform(df[['latitude', 'longitude']], sample_weight=df_labels)
```

```python
similarities[:5].round(2)
```

```python
housing_renamed = df.rename(columns={
    "latitude": "Latitude", "longitude": "Longitude",
    "population": "Population",
    "median_house_value": "Median house value (ᴜsᴅ)"})
housing_renamed["Max cluster similarity"] = similarities.max(axis=1)

housing_renamed.plot(kind="scatter", x="Longitude", y="Latitude", grid=True,
                     s=housing_renamed["Population"] / 100, label="Population",
                     c="Max cluster similarity",
                     cmap="jet", colorbar=True,
                     legend=True, sharex=False, figsize=(10, 7))
plt.plot(cluster_simil.kmeans_.cluster_centers_[:, 1],
         cluster_simil.kmeans_.cluster_centers_[:, 0],
         linestyle="", color="black", marker="X", markersize=20,
         label="Cluster centers")
plt.legend(loc="upper right")
save_fig("district_cluster_plot")
plt.show()
```

```python
from sklearn import set_config
set_config(display='diagram')
num_pipeline
```

Nous avons géré séparément les colonnes correspondant à des variables qualitatives et celles correspondant à des variables quantitatives. Il serait plus pratique d'avoir un seul transformateur capble de gérer toutes les colonnes en appliquant la transformation appropriée à chacune. 

Nous définissons la liste des noms de variables quantitatives et celle des noms de variables qualitatives et construisons un pipeline simple pour les variables qualitatives.

Ainsi, le ColumnTransformer qui suit va appliquer num_pipeline à toutes les variables quantitatives (numerical) et cat_pipeline à toutes les variables, qu'elles soient qualitatives (categorical)

```python
from sklearn.compose import ColumnTransformer

num_attribs = ["longitude", "latitude", "housing_median_age", "total_rooms",
              "total_bedrooms", "population", "households", "median_income"]
cat_attribs = ["ocean_proximity"]

cat_pipeline = make_pipeline(
    SimpleImputer(strategy="most_frequent"),
    OneHotEncoder(handle_unknown="ignore")
)

prepocessing = ColumnTransformer([
    ('num', num_pipeline, num_attribs),
    ('cat', cat_pipeline, cat_attribs),
])
```

Pour créer la liste de tous les noms de colonnes, nous utilisons la fct make_column_transformer() de Skcit-learn qui renvoie une fct de sélection nous permettant de sélectionner automatiquement toutes les variables d'un type donné, par ex les quantitatives ou les qualitatives.

```python
from sklearn.compose import make_column_selector, make_column_transformer

preprocessing = make_column_transformer(
    (num_pipeline, make_column_selector(dtype_include=np.number)),
    (cat_pipeline, make_column_selector(dtype_include=object)),
)
```

```python
# Appliquer la ColumnTransformer aux données immobilières
housing_prepared = preprocessing.fit_transform(df)
```

```python
housing_prepared_fr = pd.DataFrame(
    housing_prepared,
    columns=preprocessing.get_feature_names_out(),
    index=df.index)
housing_prepared_fr.head(2)
```

Le pipeline :
- Les valeurs manquantes des variables qualitatives seront remplacées par la médiane. Pour les variables qualitatives, les valeurs manquantes seront remplacées par la modalité la plus fréquente.
- Chaque variable qualitive subira un codage disjonctif complet car la plupart des algorithmes de ML n'acceptent que des valeurs numériques.
- Toutes les varibles numériques seront centrées et réduites, car la plupart des algorithmes de ML préfèrent que leurs variables aient à peu près la même échelle.

```python
def column_ratio(X) : 
    return X[:, [0]] / X[:, [1]]
```

```python
def ratio_pipeline():
    return make_pipeline(
        SimpleImputer(strategy="median"),
        FunctionTransformer(column_ratio, feature_names_out=ratio_name),
        StandardScaler())
```

```python
log_pipeline = make_pipeline(
    SimpleImputer(strategy="median"),
    FunctionTransformer(np.log, feature_names_out="one-to-one"),
    StandardScaler())
```

```python
cluster_simil = ClusterSimilarity(n_clusters=10, gamma=1., random_state=42)
default_num_pipeline = make_pipeline(SimpleImputer(strategy="median"),
                                     StandardScaler())
```

```python
preprocessing = ColumnTransformer([
        ("bedrooms", ratio_pipeline(), ["total_bedrooms", "total_rooms"]),
        ("rooms_per_house", ratio_pipeline(), ["total_rooms", "households"]),
        ("people_per_house", ratio_pipeline(), ["population", "households"]),
        ("log", log_pipeline, ["total_bedrooms", "total_rooms", "population",
                               "households", "median_income"]),
        ("geo", cluster_simil, ["latitude", "longitude"]),
        ("cat", cat_pipeline, make_column_selector(dtype_include=object)),
    ],
    remainder=default_num_pipeline)  # one column remaining: housing_median_age
```

### 7.2 Recherche aléatoire

La recherche par quadrillage convient bien pour explorer un nombre relativement restreint de combinaisons, comme dans l'exemeple ci-dessus, mais `RandomizedSearchCV` est souvent préalable, en particulier lorsque l'espace de recherche des hyperparamètres est vaste.

Cette classe peut être utilisée de manière similaire à la classe `GridSearchCV` mais, au lieu d'essayer toutes les combinaisons possibles, elle évalue un nombre fixé de combinaisons choisies au hasard, en sélectionnant à chaque itération une valeur aléatoire pour chaque hyperparamètre.

Cette approche a plusieurs avantages :
- Si certains de nos hyperparamètres sont continus ( ou discrets, mais avec beaucoup de valeurs possibles) et si nous laissons la recherche aléatoire effectuer 1 000 itérations par ex, alors nous explorons 1 000 valeurs différentes pour chacun de ces hyperparamètres, alors que la recherche par quadrillage explorerait uniquement les quelques valeurs indiquées pour chacun.

```python
from sklearn.experimental import enable_halving_search_cv
from sklearn.model_selection import HalvingRandomSearchCV
```

```python
# Essayer (n_iter x cv) random combinations de hyperparamètres

from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint

param_distribs = {'preprocessing__geo__n_clusters': randint(low=3, high=50),
                  'random_forest__max_features': randint(low=2, high=20)
                 }
rnd_search = RandomizedSearchCV(
    full_pipeline, param_distributions=param_distribs, n_iter=10, cv=3,
    scoring='neg_root_mean_squared_error', random_state=42
)

rnd_search.fit(df, df_labels)
```

**Afficher le résultat de random search**

```python
cv_res = pd.DataFrame(rnd_search.cv_results_)
cv_res.sort_values(by="mean_test_score", ascending=False, inplace=True)
cv_res = cv_res[["param_preprocessing__geo__n_clusters",
                 "param_random_forest__max_features", "split0_test_score",
                 "split1_test_score", "split2_test_score", "mean_test_score"]]
cv_res.columns = ["n_clusters", "max_features"] + score_cols
cv_res[score_cols] = -cv_res[score_cols].round().astype(np.int64)
cv_res.head()
```

### 7.3 Evaluer le système sur le jeu de test

```python
X_test = strat_test_set.drop("median_house_value", axis=1)
y_test = strat_test_set["median_house_value"].copy()

final_predictions = final_model.predict(X_test)

final_rmse = mean_squared_error(y_test, final_predictions, squared=False)
print(final_rmse)
```

Nous pouvons calculer un intervalle de confiance à 95% pour le test RMSE

```python
from scipy import stats

confidence = 0.95
squared_errors = (final_predictions - y_test) ** 2
np.sqrt(stats.t.interval(confidence, len(squared_errors) - 1,
                         loc=squared_errors.mean(),
                         scale=stats.sem(squared_errors)))
```

```python
m = len(squared_errors)
mean = squared_errors.mean()
tscore = stats.t.ppf((1 + confidence) / 2, df=m - 1)
tmargin = tscore * squared_errors.std(ddof=1) / np.sqrt(m)
np.sqrt(mean - tmargin), np.sqrt(mean + tmargin)
```

```python
zscore = stats.norm.ppf((1 + confidence) / 2)
zmargin = zscore * squared_errors.std(ddof=1) / np.sqrt(m)
np.sqrt(mean - zmargin), np.sqrt(mean + zmargin)
```
